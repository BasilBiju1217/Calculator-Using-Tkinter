```python
import tkinter as tk

def click(event):
    current = display.get()  # Corrected typo: diplay -> display
    text = event.widget.cget('text')  # Get the text of the clicked button

    if text == "=":  # If the "=" button is clicked
        try:
            result = eval(current)  # Evaluate the expression in the display
            display.delete(0, tk.END)  # Clear the display
            display.insert(tk.END, str(result))  # Insert the result
        except Exception as e:
            display.delete(0, tk.END)  # Clear the display
            display.insert(tk.END, "Error")  # Show "Error" if evaluation fails
    elif text == "c":  # If the "c" button is clicked
        display.delete(0, tk.END)  # Clear the display
    else:  # For all other buttons (numbers and operators)
        display.insert(tk.END, text)  # Append the button's text to the display

# Create the main window
window = tk.Tk()
window.title('Calculator')

# Create the display entry widget
display = tk.Entry(window, font=('arial', 25), justify='right')
display.pack(fill=tk.X, padx=10, pady=10)

# Create a frame to hold the buttons
btn_frame = tk.Frame(window)
btn_frame.pack()

# Define the button labels
button_labels = [
    "7", "8", "9", "/",
    "4", "5", "6", "*",
    "1", "2", "3", "-",
    "c", "0", "+", "=",
]

# Create and place the buttons in the frame
i = 0
for label in button_labels:
    button = tk.Button(btn_frame, text=label, font=("arial", 18), padx=20, pady=20)
    button.grid(row=i // 4, column=i % 4, padx=10, pady=10)  # Use grid layout
    button.bind('<Button-1>', click)  # Bind the click event to the button
    i += 1

# Start the main event loop
window.mainloop()
```
